```python
#importing all the necessary
import numpy
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input, Activation
from tensorflow.keras import layers
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import boston_housing
```

```python
print(tf.__version__)
```

```python
(X_train, y_train), (X_test, y_test) = boston_housing.load_data()
```

```python
import matplotlib.pyplot as plt
feature_set = { 'Average number of rooms' : 5}
X_train_df_1d = X_train[: , feature_set['Average number of rooms']]
X_train_df_1d
X_test_df_1d = X_test[: , feature_set['Average number of rooms']]
X_test_df_1d

plt.figure(figsize=(10, 7))
plt.xlabel('Average number of rooms')
plt.ylabel('Price')
plt.scatter(X_train_df_1d, y_train, color = 'red', label = 'Training data')
plt.scatter(X_test_df_1d, y_test, color = 'blue', label = 'Testing data')
plt.legend()
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
model = Sequential([
    Input(shape=(1,)),           # Explicit input layer
    Dense(units=1)               # Output layer with 1 unit
])

model.summary()

model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
tr = model.fit(X_train_df_1d, y_train, epochs = 100, batch_size = 15, verbose = 1)
```

```python
plt.figure(figsize=(10, 7))
plt.plot(tr.history['loss'])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```
